```python
import yfinance as yf

data = yf.download('ITC.NS', period='5y', interval='1d', group_by='columns', progress=False, ignore_tz=False)
data.insert(loc=0, column='Date', value=data.index)
data.insert(loc=0, column='Index', value=list(range(0,len(data['Open']))))
data = data.set_index('Index')
```

```python
patterns = [
 'CDLDARKCLOUDCOVER',
 'CDLDOJI',
 'CDLENGULFING',
 'CDLEVENINGDOJISTAR',
 'CDLEVENINGSTAR',
 'CDLHAMMER',
 'CDLHANGINGMAN',
 'CDLHARAMI',
 'CDLMARUBOZU',
 'CDLMORNINGDOJISTAR',
 'CDLMORNINGSTAR',
 'CDLPIERCING',
 'CDLSHOOTINGSTAR',
 'CDLSPINNINGTOP',
]
```

```python
import talib

op = data['Open']
hi = data['High']
lo = data['Low']
cl = data['Close']
for candle in patterns:
    data[candle[3:]] = getattr(talib, candle)(op, hi, lo, cl)
```

```python
import plotly.graph_objects as go
from datetime import datetime

fig = go.Figure(data=[go.Candlestick(x=data.index,
                open=op,
                high=hi,
                low=lo,
                close=cl)])
fig.show()
```
